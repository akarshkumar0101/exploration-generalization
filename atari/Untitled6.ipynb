```python
import wandb
import matplotlib.pyplot as plt
```

```python
api = wandb.Api()
```

```python
runs = list(api.runs("akarshkumar0101/egb-atari"))
runs = [run for run in runs if run.name.startswith('specialist')]
runs = [run for run in runs if run.config['obj']=='ext']
```

```python
for run in runs[:10]:
    history = run.history(samples=1000)
    plt.plot([i['charts/hns'] for i in history], label=run.config['env_ids'][0])
plt.legend()
```
