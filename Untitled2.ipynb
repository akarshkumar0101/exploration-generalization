```python
import wandb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
api = wandb.Api()
```

```python
run2df = {}
for run in tqdm(api.runs('akumar01-mit/ge_finetune_ppo')):
    if run.state != 'finished':
        continue
    run2df[run] = run.history(samples=10000)
```

```python
def calc_iqm(a, axis=0):
    assert axis==0
    a = np.sort(a, axis=0)
    n = len(a)
    low = int(np.ceil(n*.25))
    high = int(np.ceil(n*.75))
    a = a[low:high]
    return np.nanmean(a, axis=0)
```

```python
configs = [run.config for run in run2df]
```

```python
import atari_data
```

```python
aucs = []
iqms = []

plt.figure(figsize=(20, 20))
i_splits = sorted(list({config['i_split'] for config in configs}))
for i_split in i_splits:
    env_ids = sorted(list({config['env_ids'][0] for config in configs if config['i_split']==i_split}))
    for i_env_id, env_id in enumerate(env_ids):
        # print('------')
        for strategy in ['best', 'leaf']:
            runs = [run for run in run2df.keys() if run.config['i_split']==i_split and run.config['env_ids'][0]==env_id and run.config['strategy']==strategy]
            if len(runs)<4:
                continue
            # color = dict(none='blue', best='red', leaf='green')[strategy]
            color = dict(none=[0, 0, 1], best=[1, 0, 0], leaf=[0, 1, 0])[strategy]
            
            dfs = [run2df[run] for run in runs]
            scores = np.stack([df[f'returns/{env_id}_ret_score'].astype(float) for df in dfs])
            aucs.append(np.nanmean(scores, axis=-1))
            scores = atari_data.calc_hns(env_id, scores)
            # print(scores.shape)
            scores[np.isnan(scores)] = 0.
            
            
            plt.subplot(5, 8, int(i_split)*8+i_env_id+1)
            plt.title(f'{i_split} {env_id}')
            # plt.plot(np.mean(scores, axis=0), linestyle='solid', c=color, label=strategy)
            # plt.plot(np.median(scores, axis=0), linestyle='dotted', c=color, label=strategy)
            iqm = calc_iqm(scores, axis=0)
            # print(scores.shape)
            plt.plot(scores.T, linestyle='solid', c=color+[.5], label=strategy)
            # plt.plot(iqm, linestyle='solid', c=color, label=strategy)
            iqms.append(iqm.mean())
            # iqm = calc_iqm(np.nanmean(scores, axis=-1))
            # print(iqm)
            
    # plt.legend()
plt.tight_layout()
iqms = np.array(iqms).reshape(3, 8, -1)
```

```python
data = defaultdict(list)
for run in run2df.keys():
    env_id = run.config['env_ids'][0]
    strategy = run.config['strategy']
    df = run2df[run]
    score = list(df[f'returns/{env_id}_ret_score'].astype(float))
    score = np.nanmean(score)
    # score = score[-1]
    score = atari_data.calc_hns(env_id, score)
    if not np.isnan(score):
        data[strategy].append(score)

data = dict(data)
for k in data:
    data[k] = np.array(data[k])
    
```
