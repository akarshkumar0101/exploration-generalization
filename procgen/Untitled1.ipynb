```python
Il = torch.eye(2) / 0.1
Cinv = torch.zeros(1, 2, 2)  # b, d, d
Cinv[:] = Il
rews = []
for i, v in enumerate(vs.reshape(len(vs), 1, 2, 1)):
    u = Cinv @ v  # b, d, 1
    b = v.mT @ u  # b, 1, 1
    Cinv = Cinv - u @ u.mT / (1.0 + b)  # b, d, d
    rew_e3b = b[..., 0, 0].detach()
    rews.append(rew_e3b)
```
