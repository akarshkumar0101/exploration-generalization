```python
def get_batch(batch_size=128):
    env = make_env("miner", "ext", 128, 0, 0, "easy", gamma=0.999, latent_keys=[], device="cpu", cov=False, actions="ordinal")
    obs, info = env.reset()
    x, y = [obs], []
    for i in range(256):
        action = env.action_space.sample()
        obs, rew, done, info = env.step(action)
        x.append(obs)
        y.append(action)
    x = np.stack(x)
    y = np.stack(y)

    n_steps, n_envs = y.shape
    i_step, i_env = np.random.randint(0, n_steps, size=batch_size), np.arange(batch_size) % n_envs
    obs_now, obs_nxt, act_now = x[i_step, i_env], x[i_step + 1, i_env], y[i_step, i_env]
    return obs_now, obs_nxt, act_now
```

```python
def plot_grad(net):
    gs = []
    for neti in net:
        if len(list(neti.parameters()))>1:
            g = torch.cat([p.grad.flatten() for p in neti.parameters()])
            gs.append(g.abs().mean().item())
            
    plt.figure(figsize=(10, 3))
    plt.subplot(121); plt.plot(gs)
    plt.subplot(122); plt.plot(gs)
    plt.yscale('log')
```

```python
net = create_net()
net = net.to(device)
opt = torch.optim.Adam(net.parameters(), lr=3e-4)  # , weight_decay=1e-5)

losses = []
pbar = tqdm(range(1000))
for i_batch in pbar:
    obs_now, obs_nxt, act_now = get_batch(2048)

    obs_now = torch.from_numpy(obs_now).to(device)
    obs_nxt = torch.from_numpy(obs_nxt).to(device)
    act_now = torch.from_numpy(act_now).to(device)
    x = torch.cat([obs_now, obs_nxt], dim=-1).permute(0, 3, 1, 2) / 255.0
    logits = net(x)

    ce = torch.nn.functional.cross_entropy(logits, act_now, reduction="none")
    loss = ce.mean()

    opt.zero_grad()
    loss.backward()
    opt.step()

    accuracy = (logits.argmax(dim=-1) == act_now).sum().item() / len(act_now)
    data = dict(loss=loss.item(), accuracy=accuracy, ppl=np.exp(loss.item()))
    losses.append(loss.item())

    keys_tqdm = ["loss", "accuracy", "ppl"]
    pbar.set_postfix({k: data[k] for k in keys_tqdm})
    
    if i_batch%100==0:
        print(loss.item())
        plot_grad(net)
        plt.show()
```
